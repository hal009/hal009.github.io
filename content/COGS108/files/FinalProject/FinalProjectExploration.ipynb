```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import googlesearch as ggl
import requests
import bs4
```

```python
df = pd.read_csv('ngl_south_carolina.csv', index_col=False, parse_dates=['d_death_date'])
```

```python
df = df.dropna(subset=['d_first_name', 'd_last_name', 'd_birth_date', 'd_death_date'])
```

```python
df = df.loc[df['d_first_name'] != 'Unknown']
df = df.loc[df['d_birth_date'] != 'nan']

#Clean up some bad data at index 11650
df.set_value(df.index[11650], 'd_birth_date', '04/26/1940')
```

```python
df['d_birth_date'] = pd.to_datetime(df['d_birth_date'])
```

```python
df_vets = df.loc[df['relationship'] == 'Veteran (Self)']
```

```python
# Take all vets younger than 38 
df_vets = df_vets.loc[df_vets['d_birth_date'] > '1980-01-01']
drop_cols = ['d_suffix', 'section_id', 'row_num', 'site_num', 'cem_url', 'cem_phone', 'cem_addr_two', 'cem_addr_one', 'city', 'state', 'zip', 'v_first_name', 'v_mid_name', 'v_last_name', 'v_suffix']
df_vets = df_vets.drop(drop_cols, axis=1)
```
